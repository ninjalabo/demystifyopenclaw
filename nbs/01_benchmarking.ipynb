# Benchmarking

```
#| eval: false
import urllib.request
import json

def fetch_gaia_raw():
    # URL to the raw JSONL file on Hugging Face
    url = "https://huggingface.co/datasets/gaia-benchmark/GAIA/resolve/main/2023/validation/metadata.jsonl"
    
    print("Fetching raw GAIA data...")
    response = urllib.request.urlopen(url)
    
    # Parse the JSON Lines
    tasks = [json.loads(line) for line in response]
    return tasks

# Test it out
gaia_tasks = fetch_gaia_raw()
print("First GAIA Prompt:", gaia_tasks[0]['Question'])
print("First GAIA Answer:", gaia_tasks[0]['Final answer'])
```
