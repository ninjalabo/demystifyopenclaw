## Prerequisites
Ensure you have the following installed on your system:
1. Python 3.8 or higher  
2. pip (Python package installer)  
3. Access to an LLM (e.g., OpenAI's GPT-3.5 or GPT-4, or local Ollama models)  
4. Optional: A web search engine for web-based tools (default is DuckDuckGo)  

### LLM Configuration
`nanobot` is a wrapper around large language models (LLMs) that provides a simple interface for building ReAct (Reasoning + Acting) agents. To use `nanobot`, you need to have access to an LLM, such as OpenAI's GPT-3.5 or GPT-4. In this setup guide, we will use local Ollama models, but you can also configure `nanobot` to use other LLM providers.

For configuring LLM in `nanobot`, you need the following information:  
- **Model provider**: The name of the LLM provider you want to use (e.g., "ollama", "openai", etc.).  
- **Model name**: The specific model you want to use from the provider (e.g., "gpt-3.5-turbo", "gpt-4", etc.).  
- **API key**: If your LLM provider requires authentication, you will need an API key to access the model.  
- **Base URL**: The base URL for the LLM provider's API, if applicable.  

The list of supported LLM providers and their configuration details can be found in the `nanobot` documentation. Make sure to check the documentation for any specific requirements or additional configuration options for your chosen LLM provider.

### Web Search Engine Configuration
Additionally, `nanobot` includes web-based tools that require a web search engine. By default, it uses DuckDuckGo, but you can configure it to use other search engines if needed. This configuration also requires relatively similar information, including provider name, API key (if required), and base URL.

### Configuration Fields

#### 1. `agents` settings
There are several configuration fields in the `config.json` file. For our experiment which only focuses on Python and CLI commands, so sections of "channels" or "api" are not relevant. Instead, let us inspect the sections for LLM configuration, which include "agents" and "providers". Within "agents", there is a "defaults" section which includes the default configuration for all agents. We can set the LLM provider and model details here, as well as other relevant configuration fields such as temperature, max tokens, etc. The fields of relevant to our experiment include:

- `model`: The name of the provider + model you want to use (e.g., "ollama/gpt-4o", "openai/gpt-3.5-turbo", etc.).  
- `provider`: The name of the LLM provider (e.g., "ollama", "openai", etc.).  
- `temperature`: The temperature setting for the LLM, which controls the randomness of the output (lower values make the output more deterministic).  
- `maxTokens`: The maximum number of tokens to generate in the response.  
- `contextWindowTokens`: The maximum number of tokens that can be included in the context window for the LLM.  

These are the main configuration fields that influence both the performance and cost of using LLM based on baseline tasks. More extensive experiment for long-term memory / customization / multi-agent collaboration will be considered in future work.

#### 2. `providers` settings
In the "providers" section, `nanobot` supports multiple LLM providers, each with its own subsection (e.g., "openai", "ollama"), which typically includes:

- `apiKey`: The API key for authenticating with the LLM provider, if required.  
- `apiBase`: The base URL for the LLM provider's API, if applicable (e.g., for OpenAI, it would be "https://api.openai.com/v1").  
- `extraHeaders`: Any additional headers that need to be included in the API requests to the LLM provider.  

#### 3. `tools` settings
In the "tools" section, there are configuration fields for web-based tools under the "web" -> "search" key, which include:

- `provider`: The name of the web search engine provider (e.g., "duckduckgo", "google", etc.).  
- `apiKey`: The API key for authenticating with the web search engine, if required.  
- `baseUrl`: The base URL for the web search engine's API, if applicable.  
- `maxResults`: The maximum number of search results to return for each query.  
- `timeout`: The timeout setting for web search queries, in seconds.  

These configuration fields should be the basis for our experiment setup, which aims to evaluate the performance under cost constraints of nanobot agent. The following section declares the specific configuration we will use for our benchmarking task.

### Shared Experiment Configuration

````python
#| export
import json
import os
from dotenv import load_dotenv
from pathlib import Path
````

For the baseline experiment, we will use the following shared configuration for all agents:

````python
#| export
# Load secrets from the .env file into the environment
load_dotenv()

# --- LLM Provider & Model Conditions ---
LLM_PROVIDER = "ollama"
LLM_BASE_API = "http://127.0.0.1:11434/v1"
LLM_API_KEY = os.getenv("LLM_API_KEY") 
LLM_MODEL = "ministral-3:8b" 
LLM_MAX_TOKENS = 8192
LLM_CONTEXT_WINDOW = 131072
LLM_TEMPERATURE = 0.1

# --- Web Search Engine Configuration ---
WEB_SEARCH_ENGINE = "tavily"
WEB_SEARCH_BASE_URL = "https://api.tavily.com/"
WEB_SEARCH_API_KEY = os.getenv("WEB_SEARCH_API_KEY")
WEB_SEARCH_MAX_RESULTS = 5
WEB_SEARCH_TIMEOUT = 30

# --- Environment & Data ---
ENV_TASK_SET = "tasks/baseline.json"
ENV_LOG_FORMAT = "json"

DATA_DIR = "data"
DATA_LOG_DIR = os.path.join(DATA_DIR, "logs")

# --- Budget Constraints ---
ENV_API_BUDGET_EXPECTED_USD = 30
ENV_API_BUDGET_HARD_MAX_USD = 50
````

Certain variables such as `LLM_API_KEY` and `WEB_SEARCH_API_KEY` are read from the `.env` file (secrets). Configure the `.env` file with the appropriate values for your LLM provider and web search engine. For example:

```env
LLM_API_KEY=your-llm-api-key
WEB_SEARCH_API_KEY=your-web-search-api-key
```

````python
#| export
def verify_environment_variables():
    required_vars = [
        "LLM_API_KEY",
        "WEB_SEARCH_API_KEY"
    ]
    
    missing_vars = [var for var in required_vars if not os.getenv(var)]
    
    if missing_vars:
        raise EnvironmentError(f"Missing required environment variables: {', '.join(missing_vars)}")
    else:
        print("All required environment variables are set.")
````

````python
#| eval: false
verify_environment_variables()
````

We can then adjust the `config.json` file to reference these environment variables for the relevant fields. For example, in the "providers" section for the LLM provider, we can set the `apiKey` field to reference the `LLM_API_KEY` environment variable, and similarly for the web search engine configuration.

````python
#| export
def adapt_nanobot_configuration(config_file_path: str = None):
    """
    Adapt the nanobot configuration based on the current environment variables and conditions.
    
    This function can be extended to modify a configuration file or return a configuration dictionary
    that can be used by the nanobot agent during execution.
    
    Args:
        config_file_path (str, optional): Path to a configuration file to modify. If None, take the default file path.
    
    Returns:
        dict: A configuration dictionary with the adapted settings.
    """
    if config_file_path is None:
        home_dir = Path.home()
        config_file_path = home_dir / ".nanobot" / "config.json"
        if not config_file_path.exists():
            raise FileNotFoundError(f"Configuration file not found at {config_file_path}")
        
    # Load existing configuration
    with open(config_file_path, 'r', encoding='utf-8') as f:
        config = json.load(f)

    # Update default agent configuration with current environment settings
    config["agents"]["defaults"]["model"] = LLM_MODEL
    config["agents"]["defaults"]["provider"] = LLM_PROVIDER
    config["agents"]["defaults"]["contextWindowTokens"] = LLM_CONTEXT_WINDOW
    config["agents"]["defaults"]["maxTokens"] = LLM_MAX_TOKENS
    config["agents"]["defaults"]["temperature"] = LLM_TEMPERATURE

    # Update provider configuration with API keys and base URLs
    config["providers"]["ollama"]["apiKey"] = LLM_API_KEY
    config["providers"]["ollama"]["baseUrl"] = LLM_BASE_API

    # Update web search configuration
    config["tools"]["web"]["search"]["provider"] = WEB_SEARCH_ENGINE
    config["tools"]["web"]["search"]["baseUrl"] = WEB_SEARCH_BASE_URL
    config["tools"]["web"]["search"]["apiKey"] = WEB_SEARCH_API_KEY
    config["tools"]["web"]["search"]["maxResults"] = WEB_SEARCH_MAX_RESULTS
    config["tools"]["web"]["search"]["timeout"] = WEB_SEARCH_TIMEOUT

    # Save the updated configuration back to the file
    with open(config_file_path, 'w') as f:
        json.dump(config, f, indent=4)
    print(f"Configuration adapted and saved to {config_file_path}")
````

Run the `adapt_nanobot_configuration()` function to apply the current environment settings to the nanobot configuration file. If no custom configuration file is provided, it will adapt the default `config.json` file in the `.nanobot` directory.

````python
adapt_nanobot_configuration()
````
